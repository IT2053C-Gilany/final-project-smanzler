### Data set selection rationale

> Why did you select this data set?

I chose this dataset because I enjoy watching competitive Rocket League and wanted to see what I could find out from a dataset I could find.

### Questions to be answered

> Using statistical analysis and visualization, what questions would you like to be able answer about this dataset.
> This could include questions such as:
>
> - What is the relationship between X and Y variables?
> - What is the distribution of the variables?
> - What is the relationship between the variables and the target?
>   You will need to frame these questions in a way to show value to a stakeholder (i.e.why should we know about the relationship between X and Y variables?)

- What factors most strongly influence a team's chances of winning?  
  For example, is boost usage, average speed, or time spent in various field zones correlated with match outcomes? Understanding these can help teams optimize their strategies.

- How do individual player stats (e.g., shots, goals, saves) distribute across the league?  
  Knowing the typical range and distribution of these key performance indicators can help benchmark player performance and set realistic training targets.

- Are there notable differences in play style or performance between regions or teams?  
  Comparing metrics like average boost usage or defensive time by region can highlight competitive advantages and inform transfer or recruitment strategies.

## Data Loading and File I/O

This section implements robust file reading with proper exception handling.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
def load_csv(path: Path, **kwargs):
  try:
    df = pd.read_csv(path, **kwargs)
  except FileNotFoundError as e:
    raise FileNotFoundError(
        f"Could not find {path.name}. Check data/ folder and filename."
    ) from e
  except pd.errors.ParserError as e:
    raise ValueError(
        f"Parsing failed for {path.name}. Check delimiter or bad rows."
    ) from e
  else:
    return df


data_dir = Path("data")

try:
  main = load_csv(data_dir / "main.csv")
  players = load_csv(data_dir / "players_db.csv")
  games_players = load_csv(data_dir / "games_by_players.csv", low_memory=False)
  games_teams = load_csv(data_dir / "games_by_teams.csv")
  matches_players = load_csv(data_dir / "matches_by_players.csv")
  matches_teams = load_csv(data_dir / "matches_by_teams.csv")
  print("All datasets loaded successfully!")
except (FileNotFoundError, ValueError) as e:
  print(f"Error loading data: {e}")
```

## Data Preparation and Cleaning

This section prepares the data for analysis using Python data structures and pandas operations.


```python
# Convert date columns to datetime
main['match_date'] = pd.to_datetime(main['match_date'])
main['game_date'] = pd.to_datetime(main['game_date'])
main['event_start_date'] = pd.to_datetime(main['event_start_date'])

# Set categorical order for regions (using set to get unique values, then dict to map order)
unique_regions: set = set(matches_teams['team_region'].dropna())
region_order: list = sorted(list(unique_regions))  # Alphabetical order
region_cat_map: dict = {region: idx for idx, region in enumerate(region_order)}

# Convert to categorical with intentional order
matches_teams['team_region'] = pd.Categorical(
    matches_teams['team_region'],
    categories=region_order,
    ordered=True
)
matches_players['team_region'] = pd.Categorical(
    matches_players['team_region'],
    categories=region_order,
    ordered=True
)

# Create a tuple of key performance metrics for analysis
key_metrics: tuple = (
    'core_goals', 'core_shots', 'core_saves', 'core_assists',
    'boost_bpm', 'movement_avg_speed', 'advanced_rating'
)

print(
    f"Data prepared. Analyzing {len(key_metrics)} key metrics across regions.")

# Merge match data with main to get dates and event info
matches_teams_enriched = matches_teams.merge(
    main[['match_id', 'match_date', 'event_region', 'event_split']].drop_duplicates(),
    on='match_id',
    how='left'
)

# Merge player match data with player database
matches_players_enriched = matches_players.merge(
    players[['player_id', 'player_tag', 'player_country']],
    on='player_id',
    how='left',
    suffixes=('', '_db')
)

# Use the tag from database if available, otherwise use the one from matches
matches_players_enriched['player_tag'] = matches_players_enriched['player_tag_db'].fillna(
    matches_players_enriched['player_tag']
)

print(
    f"Merged datasets: {matches_teams_enriched.shape[0]} team matches, {matches_players_enriched.shape[0]} player matches")

# Groupby with aggregation: Average team stats by region
team_stats_by_region = matches_teams_enriched.groupby('team_region', observed=True).agg({
    'core_goals': ['mean', 'std'],
    'core_shots': ['mean', 'std'],
    'core_saves': ['mean', 'std'],
    'boost_bpm': ['mean', 'std'],
    'movement_total_distance': ['mean', 'std'],
    'winner': 'mean'
}).round(2)

team_stats_by_region.columns = [
    '_'.join(col).strip() for col in team_stats_by_region.columns.values]
print("Team statistics by region:")
print(team_stats_by_region)

# Tidy reshape: Pivot table for player performance by region
player_perf_pivot = matches_players_enriched.pivot_table(
    values=['core_goals', 'core_saves', 'advanced_rating'],
    index='team_region',
    aggfunc='mean',
    observed=True
).round(2)

print("Player performance pivot table by region:")
print(player_perf_pivot)
```

### Visualization 1: Distribution plot - Goals per player

This plot shows the distribution and summary stats of goals scored per player per match using a histogram.


```python
# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Visualization 1: Distribution plot - Goals per player (histogram only)
fig, ax = plt.subplots(figsize=(10, 6))
player_goals = matches_players_enriched['core_goals'].dropna()

max_goals = int(player_goals.max())
bins = [x - 0.5 for x in range(max_goals + 2)]

# Histogram with counts (not density) - shows actual number of players
ax.hist(player_goals, bins=bins, density=False, alpha=0.7,
        color='steelblue', edgecolor='black')

ax.set_xlabel('Goals per Match', fontsize=12)
ax.set_ylabel('Frequency (Count)', fontsize=12)
ax.set_title('Distribution of Goals per Player per Match',
             fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Summary statistics
print(f"\nGoals Distribution Summary:")
print(f"Mean: {player_goals.mean():.2f}")
print(f"Median: {player_goals.median():.2f}")
print(f"Std: {player_goals.std():.2f}")
print(f"Skew: {player_goals.skew():.2f}")
print(f"Min: {player_goals.min():.2f}, Max: {player_goals.max():.2f}")
```

### Visualization 2: Comparison plot - Box plot of boost usage by region

Box plot comparing boost per minute (boost_bpm) by region.


```python
# Visualization 2: Comparison plot - Box plot of boost usage by region
fig, ax = plt.subplots(figsize=(12, 6))

# Filter out any rows with missing data for region or boost_bpm
data_for_box = matches_teams_enriched.dropna(
    subset=['team_region', 'boost_bpm'])

sns.boxplot(
    data=data_for_box,
    x='team_region',
    y='boost_bpm',
    hue='team_region',
    legend=False,
    ax=ax,
    palette='Set2'
)

# Remove duplicate region labels from the legend if present
handles, labels = ax.get_legend_handles_labels()
if len(handles) > 0:
  ax.get_legend().remove()

ax.set_xlabel('Region', fontsize=12)
ax.set_ylabel('Boost Used per Minute', fontsize=12)
ax.set_title('Boost Usage Comparison Across Regions',
             fontsize=14, fontweight='bold')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
